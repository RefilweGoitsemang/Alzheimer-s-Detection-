```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model('/content/drive/MyDrive/alzheimers_detection_model.h5')
```

```python
import os
import cv2
import numpy as np
# Function to preprocess the image
def preprocess_image(image_path):
    img = cv2.imread(image_path)  # Read the image
    img = cv2.resize(img, (128, 128))  # Resize to 128x128
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    img = img.reshape(128, 128, 1)  # Update shape to 128x128x1
    img = img / 255.0  # Normalize pixel values to [0, 1]
    return np.expand_dims(img, axis=0)  # Add batch dimension

# Path to the image
image_path = '/content/drive/MyDrive/train/ModerateDemented/moderateDem41.jpg'

# Preprocess the image
input_image = preprocess_image(image_path)

```

```python
# Define class labels
class_labels = {
    0: "Non Demented",
    1: "Very Mild Demented",
    2: "Mild Demented",
    3: "Moderate Demented"
}

# Predict the class
predictions = model.predict(input_image)
class_index = np.argmax(predictions)  # Get the index of the highest probability

print(f"Predicted Class: {class_index}")
```
